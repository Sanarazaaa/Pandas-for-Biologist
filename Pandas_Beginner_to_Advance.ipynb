```python
import pandas as pd

# Create a sample DataFrame (or load your data)
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
df = pd.DataFrame(data)

# Now, check for duplicate indices:
df.index.is_unique
```

```python
# return a boolean ndarray indicating whether a label is repeated.

df.index.duplicated()
```

```python
# filter to drop duplicate rows

df.loc[~df.index.duplicated(), :]
```

```python
# Get the 'City' of the second row
city_bob = df.loc[1, 'City']
print(city_bob)  # Output: Los Angeles
```

```python
df.groupby(level=0)[['Age']].mean() # Only calculate the mean for the 'Age' column
# mean is only calculated for numeric
```

```python
# Inconsistent dataframes can make merging datasets harder. We can standardize formats and for this, use .lower() for text and .strftime() for dates.

data = {'Name': ['ALICE', 'bob', 'ChArLiE']}
df = pd.DataFrame(data)  # Create a DataFrame from the data dictionary
df['Name'] = df['Name'].str.lower() # Convert to lowercase
df
```

```python
df = pd.DataFrame(data)

# Add a datetime column
df['Experiment_Date'] = pd.date_range(start='2023-01-01', periods=len(df), freq='D')

# Format the dates using strftime
df['Formatted_Date'] = df['Experiment_Date'].dt.strftime('%d-%b-%Y')

print(df)
```

# Identify Outliers

**Statistical Methods**

**Z-Score:** Outliers are identified if the z-score is greater than a threshold (commonly 3).

**IQR (Interquartile Range):**
Outliers are defined as values outside the range [Q1 - 1.5 * IQR, Q3 + 1.5 * IQR].

**Custom Threshold:**
Define specific thresholds for your dataset based on domain knowledge.
